```python
import numpy as np

X = np.array([[0,0], [0,1], [1,0], [1,1]])
y = np.array([0, 0, 0, 1])
```

 W_j = W_j + LR * Err * I_j

```python
def train_perceptron(X, y, lr=0.1, epochs=10):
    weights = np.zeros(X.shape[1])
    bias = 0
    
    for epoch in range(epochs):
        for i in range(len(X)):
            net = np.dot(X[i], weights) + bias
            
            prediction = 1 if net > 0 else 0
            
            error = y[i] - prediction
            
            weights += lr * error * X[i]
            bias += lr * error
            
    return weights, bias

weights, bias = train_perceptron(X, y)
print(f"Final Weights: {weights}, Bias: {bias}")
```

```python
def predict(credit, income):
    net = (credit * weights[0]) + (income * weights[1]) + bias
    return "Approved" if net > 0 else "Rejected"

print(predict(1, 1)) 
```
